Import pi from math modules

```python
from math import pi
```

Calculate power when rpm and Torque is given using this equation

`n = 1800 rpm`

`Torque = 0.8 N.m` 

$Power, P= \frac{2 \pi n}{60} \times Torque$

```python
n=1800
Torque = 0.8
Power = Torque * n * 2 * pi / 60
print('Power =', Power, 'W')
```

Rounding the Power to 3 decimals

```python
Power = round(Torque * n * 2 * pi / 60,2)
print('Power =', Power, 'W')
```
