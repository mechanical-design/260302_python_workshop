#### Task 2
Calculate power when rpm and Torque is given using this equation

`n = 1800 rpm`

`Torque = 0.8 N.m` 

$Power, P= \frac{2 \pi n}{60} \times Torque$

```python
n=1800
Torque = 0.8
Power = Torque * n * 2 * pi / 60
print('Power =', Power, 'W')
```

#### Task 3
Rounding the Power to 3 decimals

Example:

`Power = round(Torque * n * 2 * pi / 60,2)`

Output:

`Power = 150.8 W`

```python
Power = round(Torque * n * 2 * pi / 60,3)
print('Power =', Power, 'W')
```

#### Task 4
A solid square rod with diameter, d=5mm is pulled along the axis with an equivalent force of Force = (300 cos 15)N. Calculate the stress using this equation.

$\sigma = \dfrac{Force}{Area}$

```python
from math import cos, radians
```

```python
d = 0.005
Force = 300 * cos(radians(15))
Area = pi * d**2/4
stress = Force / Area
print('Stress =', stress, 'Pa')
```

#### Task 5
Convert the Stress to MPa

```python
print(f'Stress = {stress/1e6:.2f} MPa')
# or
# print('Stress = {:.2f} MPa'.format(stress/1e6))
```

#### Task 6
Convert the Stress to Engineering notations

```python
print(f'Stress = {stress/1e6:.2e} MPa')
```
